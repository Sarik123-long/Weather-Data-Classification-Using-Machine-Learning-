```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
```

```python
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/_0eYOqji3unP1tDNKWZMjg/weatherAUS-2.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.count()
```

```python
df = df.dropna()
df.info()
```

```python
df.columns
```

```python
df = df.rename(columns={'RainToday': 'RainYesterday',
                        'RainTomorrow': 'RainToday'
                        })
```

```python
df = df[df.Location.isin(['Melbourne','MelbourneAirport','Watsonia',])]
df. info()
```

```python
def date_to_season(date):
    month = date.month
    if (month == 12) or (month == 1) or (month == 2):
        return 'Summer'
    elif (month == 3) or (month == 4) or (month == 5):
        return 'Autumn'
    elif (month == 6) or (month == 7) or (month == 8):
        return 'Winter'
    elif (month == 9) or (month == 10) or (month == 11):
        return 'Spring'
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Apply the function to the 'Date' column
df['Season'] = df['Date'].apply(date_to_season)

df=df.drop(columns=['Date'])
df
```

```python
X = df.drop(columns='RainToday', axis=1)
y = df['RainToday']
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
numeric_features = X_train.select_dtypes(include=['number']).columns.tolist()
categorical_features = X_train.select_dtypes(include=['object']).columns.tolist()
```

```python
# Scale the numeric features
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])

# One-hot encode the categoricals
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
param_grid = {
    'classifier__n_estimators': [50, 100],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5]
}
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
print(X_train_processed.shape)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
pipeline = Pipeline(steps=[
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
from sklearn.model_selection import GridSearchCV

# Assuming you already have your model and param_grid
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5)

# Fit the model
grid_search.fit(X_train_processed, y_train)

# Now you can safely access these
print("\nBest parameters found: ", grid_search.best_params_)
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

```python
test_score = grid_search.score(X_test_processed, y_test)
print("Test set score: {:.2f}".format(test_score))
```

```python
y_pred = grid_search.predict(X_test_processed)
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
feature_importances = grid_search.best_estimator_['classifier'].feature_importances_
print("Feature Importances:")
print(feature_importances)
```

```python
# Get feature names after preprocessing
# Numerical features remain the same
numeric_feature_names = numeric_features

# Categorical features are one-hot encoded
# Need to get the feature names from the one-hot encoder within the preprocessor
onehot_features = preprocessor.named_transformers_['cat'].named_steps['onehot']
categorical_feature_names = onehot_features.get_feature_names_out(categorical_features)

# Combine numerical and categorical feature names in the correct order
feature_names = list(numeric_feature_names) + list(categorical_feature_names)

feature_importances = grid_search.best_estimator_['classifier'].feature_importances_

importance_df = pd.DataFrame({'Feature': feature_names,
                              'Importance': feature_importances
                             }).sort_values(by='Importance', ascending=False)

N = 20  # Change this number to display more or fewer features
top_features = importance_df.head(N)

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(top_features['Feature'], top_features['Importance'], color='skyblue')
plt.gca().invert_yaxis()  # Invert y-axis to show the most important feature on top
plt.title(f'Top {N} Most Important Features in predicting whether it will rain today')
plt.xlabel('Importance Score')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Redefine the pipeline with LogisticRegression
# Note: We are using the preprocessor from earlier steps, but we are
# fitting the grid search on already preprocessed data, so the pipeline
# only needs the classifier step.
pipeline_lr = Pipeline(steps=[
    ('classifier', LogisticRegression(random_state=42))
])

# 2. Define a new grid with Logistic Regression parameters, ensuring compatibility
param_grid_lr = [
    {
        'classifier__solver': ['liblinear'],
        'classifier__penalty': ['l1', 'l2'],
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__class_weight': [None, 'balanced']
    },
    {
        'classifier__solver': ['lbfgs'],
        'classifier__penalty': ['l2'], # lbfgs only supports l2
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__class_weight': [None, 'balanced']
    }
]

# 3. Create a new GridSearchCV object
# Use the preprocessed training data X_train_processed
grid_search_lr = GridSearchCV(estimator=pipeline_lr, param_grid=param_grid_lr, cv=5, scoring='accuracy', verbose=2)

# 4. Fit the new GridSearchCV object
grid_search_lr.fit(X_train_processed, y_train)

# Print best parameters and score for Logistic Regression
print("\nLogistic Regression Best parameters found: ", grid_search_lr.best_params_)
print("Logistic Regression Best cross-validation accuracy: {:.2f}".format(grid_search_lr.best_score_))

# 5. Evaluate the best Logistic Regression model on the test set
test_score_lr = grid_search_lr.score(X_test_processed, y_test)
print("Logistic Regression Test set accuracy: {:.2f}".format(test_score_lr))

# Make predictions for evaluation metrics
y_pred_lr = grid_search_lr.predict(X_test_processed)

# Classification Report
print("\nLogistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))

# Confusion Matrix
conf_matrix_lr = confusion_matrix(y_test, y_pred_lr)
disp_lr = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_lr)
disp_lr.plot(cmap='Blues')
plt.title('Logistic Regression Confusion Matrix')
plt.show()
```
